```python
import requests
import re
import pandas as pd
```

```python
def get_revisions(article_title,names, year,lang):
    revisions = []
    url='https://'+lang+'.wikipedia.org/w/api.php?action=query&list=search&srsearch='+article_title+'&format=json'
    #setting start time as the first day of the year and end time as the last day of the year
    start_time = year + "-01-01T00:00:00Z"
    end_time = year + "-12-31T23:59:59Z"
    params = {
        "action": "query",
        "format": "json",
        "prop": "revisions",
        "titles": article_title,
        "rvlimit": 1,  #Get one revision 
        "rvprop": "content", #get time stamp and content
        "rvdir": "newer",  #Retrieve newer revisions first
        "rvstart": start_time,
        "rvend": end_time
    }
    try:
        #open the url and get the values
        response = requests.get(url, params=params)
        data = response.json()
        pages = data['query']['pages']
        page = next(iter(pages.values()))  

        #if there is a revision, get the content
        if 'revisions' in page:
            content=page['revisions'][0]['*']
            #extract all links in the content
            links=re.findall(r'\[\[(.*?)\]\]', content)
            #find the names that are in the list of scientistsnames
            links=[link for link in links if link in names]
            #combine all information in a dictionary
            revision = {'Name': article_title, 'Year': year, 'Links': links}
            revisions.append(revision)
    except:
        pass #nothing is done if the page is not found
    #convert to a dataframe
    revisions_df = pd.DataFrame(revisions)
    return revisions_df
```

```python
scientists = pd.read_csv('Data/titles.csv')
scientistsnames=scientists['Name'].unique().tolist()
n=len(scientistsnames)
years=['2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021','2022','2023','2024']
c=0
language='en'

#creating dataframes to store the data and save it to csv

#df=pd.DataFrame(columns=['Name','Year','Links'])
#df_names=pd.DataFrame(columns=['Name'])
#df.to_csv('revisions_en.csv',index=False)
#df_names.to_csv('names_en.csv',index=False)

while c<n:
    df=pd.read_csv('revisions_en.csv')
    names=pd.read_csv('names_en.csv')['Name'].tolist()
    missing = [name for name in scientistsnames if name not in names]
    for name in missing[0:10]:
        names.append(name)
        for year in years:
            revisions = get_revisions(name,scientistsnames,year,language)
            df = pd.concat([df,revisions],ignore_index=True)
            df_names=pd.DataFrame(names,columns=['Name'])
    df.to_csv('revisions_en.csv',index=False)
    df_names.to_csv('names_en.csv',index=False)
    missing = [name for name in scientistsnames if name not in names]
    c=len(names) 
```
