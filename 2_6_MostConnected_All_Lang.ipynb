```python
import pandas as pd
import get_dataremodelled as gdr
import graphcreating as gc
```

```python
def get_degree(G,community,df_community):
    indegree=[]
    outdegree=[]
    degree=[]
    names=[]
    members=df_community[df_community['Community ID']==community]
    members=members['Members'].tolist()
    for node in G.nodes:
        if node in members:
            names.append(node)
            indegree.append(G.in_degree(node))
            outdegree.append(G.out_degree(node))
            degree.append(G.degree(node))
    df=pd.DataFrame({'Name':names,'InDegree':indegree,'OutDegree':outdegree,'Degree':degree, 'Community':community})
    return df
```

```python
gender=pd.read_csv('gendered_titles_final.csv')
gender=gender.drop_duplicates(subset='Name')
df_names=pd.read_csv('Data/titles.csv')
```

```python
lang=''
df_community=pd.read_csv('all_communities_infomap'+lang+'.csv',sep=';')
df_community['Members']=df_community['Members'].apply(lambda x: x.split(', '))
df_community=df_community.explode('Members')
df_community=df_community[df_community['Year']==2024]

lang='en'
focuscommunities=[1,2,3,4,8]

revisions=pd.read_csv('Data/revisions_'+lang+'.csv')
revisions=gdr.prep_data(revisions,gender)
revisions=gdr.prep_links(revisions,gender)
revisions=revisions[revisions['Year']==2024]

G=gc.create_directed_graph(revisions,gender)
```

```python
for com in focuscommunities:
    df=get_degree(G,com,df_community)
    df=df.sort_values(by='InDegree',ascending=False)
    df=df.merge(gender,how='left',on='Name')
    print('Community:',com)
    print('Average InDegree:',df['InDegree'].mean())
    print('Average OutDegree:',df['OutDegree'].mean())
    print('Top 10 InDegree:')
    list=df['Name'].tolist()
    genderlist=df['gender'].tolist()
    print(list[:10])
    print('Top 10 InDegree gender:')
    print(genderlist[:10])
    print('')
```
