```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_en=pd.read_csv('all_communities_infomap.csv',sep=";")
df_en=df_en[df_en['Year']==2024]
df_es=pd.read_csv('all_communities_infomap_es.csv',sep=";")
df_es=df_es[df_es['Year']==2024]
df_fr=pd.read_csv('all_communities_infomap_fr.csv',sep=";")
df_fr=df_fr[df_fr['Year']==2024]
df_de=pd.read_csv('all_communities_infomap_de.csv',sep=";")
df_de=df_de[df_de['Year']==2024]
```

```python

df_fr['Community ID'] = df_fr['Community ID'].astype(str)
df_fr = df_fr[df_fr['Size'] > 50].sort_values(by='Size')

df_es['Community ID'] = df_es['Community ID'].astype(str)
df_es = df_es[df_es['Size'] > 50].sort_values(by='Size')

df_de['Community ID'] = df_de['Community ID'].astype(str)
df_de = df_de[df_de['Size'] > 50].sort_values(by='Size')

df_en['Community ID'] = df_en['Community ID'].astype(str)
df_en = df_en[df_en['Size'] > 50].sort_values(by='Size')

x_fr = df_fr['Community ID'].tolist()
size_fr = df_fr['Percentage of Women'].tolist()
y_fr = df_fr['Size'].tolist()

x_es = df_es['Community ID'].tolist()
size_es = df_es['Percentage of Women'].tolist()
y_es = df_es['Size'].tolist()

x_de = df_de['Community ID'].tolist()
size_de = df_de['Percentage of Women'].tolist()
y_de = df_de['Size'].tolist()

x_en = df_en['Community ID'].tolist()
size_en = df_en['Percentage of Women'].tolist()
y_en = df_en['Size'].tolist()

color1='#6DBE98'
color2='#436436'

fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Plot for French communities
ax1 = axs[0, 0]
bars=ax1.bar(x_fr, y_fr, color=color1, width=0.4, label='Percentage of Women')
ax1.set_xlabel("Community ID")
ax1.set_ylabel("Community Size")
ax1.set_title("French")
ax1.tick_params(axis='x', rotation=90)
ax1.grid(axis='y', which='both')


ax1_twin = ax1.twinx()
bars_size=ax1_twin.bar(x_fr, size_fr, color=color2, width=0.2, alpha=1, label='Size')
ax1_twin.set_ylim(0,25)
ax1_twin.set_ylabel("Percentage female members")

# Plot for Spanish communities
ax2 = axs[0, 1]
ax2.bar(x_es, y_es, color=color1, width=0.4, label='Percentage of Women')
ax2.set_xlabel("Community ID")
ax2.set_ylabel("Community Size")
ax2.set_title("Spanish")
ax2.tick_params(axis='x', rotation=90)
ax2.grid(axis='y', which='both')

ax2_twin = ax2.twinx()
ax2_twin.bar(x_es, size_es, color=color2, width=0.2, alpha=1, label='Size')
ax2_twin.set_ylim(0,25)
ax2_twin.set_ylabel("Percentage female members")

# Plot for German communities
ax3 = axs[1, 0]
ax3.bar(x_de, y_de, color=color1, width=0.4, label='Percentage of Women')
ax3.set_xlabel("Community ID")
ax3.set_ylabel("Community Size")
ax3.set_title("German")
ax3.tick_params(axis='x', rotation=90)
ax3.grid(axis='y', which='both')

ax3_twin = ax3.twinx()
ax3_twin.bar(x_de, size_de, color=color2, width=0.2, alpha=1, label='Size')
ax3_twin.set_ylim(0,25)
ax3_twin.set_ylabel("Percentage female members")

# Plot for English communities
ax4 = axs[1, 1]
ax4.bar(x_en, y_en, color=color1, width=0.4, label='Percentage of Women')
ax4.set_xlabel("Community ID")
ax4.set_ylabel("Community Size")
ax4.set_title("English")
ax4.tick_params(axis='x', rotation=90)
ax4.grid(axis='y', which='both')

ax4_twin = ax4.twinx()
ax4_twin.bar(x_en, size_en, color=color2, width=0.2, alpha=1, label='Size')
ax4_twin.set_ylim(0,25)

ax4_twin.set_ylabel("Percentage female members")


fig.legend([bars, bars_size], ['Size','Percentage female members'], loc='lower center', ncol=2)

plt.tight_layout()
plt.show()
```
